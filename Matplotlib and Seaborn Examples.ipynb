# Matplotlib and Seaborn Reference Examples

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Matplotlib

## Basic plotting function

```python
x = np.arange(0,10)
y = 2*x
```

```python
plt.plot(x,y)
plt.title("Example Function")
plt.xlim(0,9)
plt.ylim(0,18)
plt.xlabel("x")
plt.ylabel("y");
#plt.savefig("Example1.png")
```

## Figure Object

```python
# Define data - linearly spaced points and range of integers spaced by one
a = np.linspace(0,10,11)
b = a**4
x = np.arange(0,10)
y = 2*x
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b);
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y);
```

```python
# How to add an insert axes of zoomed image

fig = plt.figure(dpi = 80,figsize = (6,4))
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.15,0.6,0.25,0.25])
axes1.plot(a,b)
axes2.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("Matplotlib Objects")

axes2.set_xlim(1,2)
axes2.set_ylim(0,20)
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("Zoomed In")

plt.show()

#fig.savefig("New_figure.png",bbox_inches = "tight")
```

## Subplots Functionality

```python
fig, axes = plt.subplots(nrows = 1,ncols = 2)
```

```python
fig, axes = plt.subplots(nrows = 2,ncols = 1)
```

```python
fig, axes = plt.subplots(nrows = 2,ncols = 2)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(x,y)
axes[1].plot(a,b)
plt.tight_layout();
```

```python
fig, axes = plt.subplots(nrows = 2,ncols = 2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(x,y)
plt.tight_layout();
```

```python
fig, axes = plt.subplots(nrows = 3,ncols = 1)

for ax in axes:
    ax.plot(x,y)
    
plt.tight_layout()
```

```python
# Adjust spacing manually without using tight layout option
fig, axes = plt.subplots(nrows = 2,ncols = 2, figsize = (10,4))
axes[0][0].plot(x,y) # can set titles, axes labels etc for each one also
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(x,y)

fig.suptitle("Comparison of All Plots", size = 15, y = 0.99) # Can position the title
plt.tight_layout()
fig.set_figwidth = (10)
fig.subplots_adjust(wspace = 0.5, hspace = 0.5) # can also adjust left, right, top and bottom
```

## Styling

```python
fig, ax = plt.subplots()
ax.plot(x,x, label = "X")
ax.plot(x,x**2, label = "X to power 2")
ax.legend(loc = "upper center") 
ax.set_title("Styling", fontsize = 20);
```

```python
# Put legend outside the plot use tuple 
fig, ax = plt.subplots()
ax.plot(x,x, label = "X")
ax.plot(x,x**2, label = "X to power 2")
ax.legend(loc = (1.1,0.5)) 
ax.set_title("Styling", fontsize = 20);
```

```python
fig, ax = plt.subplots()
ax.plot(x,x, label = "X",color = "purple")
ax.set_title("Styling", fontsize = 20);
```

```python
# Change color with hex code, linewidth
fig, ax = plt.subplots()
ax.plot(x,x, label = "X",color = '#8803fc',linewidth = 3, linestyle = "--") # can create custom linestyles
ax.set_title("Styling", fontsize = 20);
```

```python
# Create custom linestyle
fig, ax = plt.subplots()
lines = ax.plot(x,x, label = "X",color = '#8803fc',linewidth = 3) 
lines[0].set_dashes([1,2,1,2,10,2])
ax.set_title("Styling", fontsize = 20);
```

```python
# Change markers
fig, ax = plt.subplots()
ax.plot(x,x, label = "X",color = 'green',linewidth = 2, marker = "o",markersize = 10, 
        markerfacecolor = "blue",markeredgewidth = 4, markeredgecolor = "red") # many different marker types
ax.set_title("Styling", fontsize = 20);
```

## Scatter Plots

```python
# Use matplotlib and seaborn using hue set to categorical column
plt.figure(figsize = (10,5),dpi = 75)
plt.title("Salary Against Sales")
sns.scatterplot(x = "salary", y = "sales", data = df, alpha = 0.5, hue = "level of education", palette = "Dark2");
```

```python
# Use matplotlib and seaborn using hue set to salary (adds a gradient)
plt.figure(figsize = (10,5),dpi = 75)
plt.title("Salary Against Sales")
sns.scatterplot(x = "salary", y = "sales", data = df, alpha = 0.5, hue = "salary");
```

```python
# Use matplotlib and seaborn using size parameter to show the years of work experience (s = size of points)
plt.figure(figsize = (10,5),dpi = 75)
plt.title("Salary Against Sales")
sns.scatterplot(x = "salary", y = "sales", data = df, alpha = 0.5, size = "work experience", s = 150);
```

```python
# Style parameter (use style and colour for the level of education)
plt.figure(figsize = (10,5),dpi = 75)
plt.title("Salary Against Sales")
sns.scatterplot(x = "salary", y = "sales", data = df, alpha = 0.5, size = "level of education", 
                style = "level of education", hue = "level of education");
# plt.savefig("Seaborn Scatter.png")
```

## Distribution Plots

```python
df.head()
```

```python
# Salary distributions
plt.figure(figsize = (8,5))
sns.rugplot(x = "salary", data = df, height = 0.5);
```

```python
# Distplot (can also use histplot)
sns.set(style = "ticks") # also whitegrid, white, dark, darkgrid
sns.displot(data = df, x = "salary",bins = 25, color = "lightsteelblue",edgecolor = "black", 
            kde = True,linewidth = 1)
plt.title("Distplot and Kde for Salary Data")
plt.figure(figsize = (10,6));
```

```python
# KDE plot
sns.kdeplot(x = "salary", data = df);
```

```python
# Focus on a section of the plot
sns.kdeplot(x = "salary", data = df, clip = [80000,100000],shade = True);
```

## Pair Grid

```python
g = sns.PairGrid(df, hue = "gender")
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot) 
g = g.add_legend() # also kdeplot gives the normal plot for kde
```

## Facet Grid

```python
g = sns.FacetGrid(df, col="gender",  row="lunch")
g.map(sns.scatterplot, "math score", "reading score")
g = g.add_legend()
```
